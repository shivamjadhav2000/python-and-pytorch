```python
def maxArea( height):
        l,r=0,len(height)-1
        volumes=[]
        while l<r:
            print(f"l={l},r={r},volumes={volumes},height[l]={height[l]},height[r]={height[r]}")
            distance=r-l
            dif=min(height[l],height[r])
            volumes.append(dif*distance)
            if height[l]>height[r]:
                r-=1
            else:
                l+=1
        return max(volumes)
```

```python
def checkparanthesis(l):
    d={'(':')','{':'}','[':']',')':'(','}':'{',']':'['}
    history={']':0,'}':0,')':0}
    starts='({['
    ends='})]'
    stack=[]
    for i in l:
        stack.append(i)
        if i in ends:
            if history[i]<=0:
                return False
            if len(stack)>1 and d[i]==stack[-2]:
                stack=stack[:-2]
                history[i]-=1
        else:
            history[d[i]]=max(1,history[d[i]]+1)
    return False if len(stack) else True
        
        
        
```

```python
def removeDuplicates( nums):
        i=1
        if len(nums)<=1:
            return nums
        for _ in range(1,len(nums)):
            if nums[i] in nums[:i]:
                nums.pop(i)
            else:
                i+=1
        return nums
```
